```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('../data/Salary_Data.csv')
data
```

```python
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=0)
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(X_train, y_train)
```

```python
vt = plt
vt.scatter(X_train, y_train, color='red')
vt.plot(X_train, regr.predict(X_train), color='blue')
vt.title('Salários vs Tempo de experiência')
vt.xlabel('Anos de trabalho')
vt.ylabel('Valor do salário')
vt.show()
```

```python
vt = plt
vt.scatter(X_test, y_test, color='red')
vt.plot(X_test, regr.predict(X_test), color='blue')
vt.title('Salários vs Tempo de experiência')
vt.xlabel('Anos de trabalho')
vt.ylabel('Valor do salário')
vt.show()
```

```python
print('Slop: ', regr.coef_)
print('Intercept: ', regr.intercept_)

## f(x) = 9360.26 X + 26777.39
```

```python
## Precisão/Qualidade
y_pred = regr.predict(X_test)
print('MSE :', mean_squared_error(y_test, y_pred))
print('R2: ', r2_score(y_test, y_pred))
```
